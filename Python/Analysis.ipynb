```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load CSV (the query export)
df = pd.read_csv("slow_movers.csv")

# Quick check
print(df.head())
print(df.info())

# Correlation coefficient
corr = df['days_since_last_sale'].corr(df['profit_margin_pct'])
print("Correlation between Days Since Last Sale and Profit Margin (%):", corr)
#Interpretation:
#Close to 0 → no strong relation.
#Negative → more days since last sale → lower profit margin (likely dead stock).
#Positive → unusual; means old stock has high margin (rare).

plt.figure(figsize=(8,6))
sns.regplot(x='days_since_last_sale', y='profit_margin_pct', data=df, scatter_kws={'alpha':0.6})
plt.xlabel("Days Since Last Sale")
plt.ylabel("Profit Margin (%)")
plt.title("Correlation: Days Since Last Sale vs Profit Margin")
plt.show()
```
